### 2. ###
You now need to derive class labels for each of the data points in the training data. You can do this automatically using the K-means algorithm applied to `X`, where *K* is the number of classes you identified in the previous step. Use the function `kmeans` to do this. Next, you want to visualise the outcome of the K-means clustering. One way to do this is to plot each cluster found by K-means in a different colour. To this end, use the vector of cluster indices returned by `kmeans` to store the points for each class in separate matrices by means of the `np.where` command.

```python
def kmeans(Data, NClusters):
  km = KMeans(NClusters)
  fitted = km.fit(Data)
  return (fitted.cluster_centers_, fitted.labels_, fitted.inertia_)

clustercentres, labels, inertias = kmeans(data, 3)

fig0 = plt.figure()
plt.title('Features 1 and 2 with three clusters and their centroids')
ax0 = fig0.add_subplot( 111)

def clusterFromLabels(labels, data):

    # Points from cluster 1
    indicies1 = np.where(labels == 0)
    points1 = data[indicies1, :]
    
    # Points from cluster 2
    indicies2 = np.where(labels == 1)
    points2 = data[indicies2, :]
    
    # Points from cluster 3
    indicies3 = np.where(labels == 2)
    points3 = data[indicies3, :]
    
    return (points1, points2, points3)

def plotClusters(points1, points2, points3, ax, mark, size, datatype):
    
    #Plot Cluster 1 
    ax.scatter(points1[0][:,0], points1[0][:,1], marker=mark, s= size, label = datatype + "Cluster1")

    #Plot Cluster 2
    ax.scatter(points2[0][:,0], points2[0][:,1], marker=mark, s = size, label = datatype +  "Cluster2")

    #Plot Cluster 3
    ax.scatter(points3[0][:,0], points3[0][:,1], marker=mark, s = size, label = datatype + "Cluster3")
    
    #Show the legends
    ax.legend()

#Add cluster centres to plot
ax0.scatter(clustercentres[:,0], clustercentres[:,1], marker="+", s=900, label= "Cluster Centroids")    
    
ps1, ps2, ps3 = clusterFromLabels(labels, data)    
plotClusters(ps1, ps2, ps3, ax0, "o", 100, "Training Data ")








```

### 3. ###
The centroids found by K-means can now be used as a simple nearest-neighbour classifier. Load your test data and select the two relevant features. Using a combination of the `cdist(X, metric='euclidean', p=2)` and `np.argmin` (with `axis` parameter) commands, construct a vector of labels for each of the test points indicating which cluster centroid is the nearest. Plot the test points in the previous plot with the colour indicating their class but a different symbol. The corresponding decision boundaries can be plotted by passing the centroids to the `Voronoi` and `voronoi_plot_2d` functions (remember to set the `ax` parameter of the later one to your current plot).

```python
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)

vor = Voronoi(clustercentres)
fig10 = voronoi_plot_2d(vor)
plt.title('Clustered training data with test data and decision boundaries')

ax1 = fig10.add_subplot(111)
ax1.set_xlim(1, 10)
ax1.set_ylim(1, 10)

ps1, ps2, ps3 = clusterFromLabels(labels, data)    
plotClusters(ps1, ps2, ps3, ax1, "o", 100, "Training Data ")

#Add cluster centres to plot
ax1.scatter(clustercentres[:,0], clustercentres[:,1], marker="+", s=900, label= "Cluster Centroids")
    
test = np.loadtxt("nb16568.test")
testData = test[:,[1, 2]]

dist = cdist(clustercentres, testData, metric='euclidean', p=2)
indicies = np.argmin(dist, axis=0)


testPoints1, testPoints2, testPoints3 = clusterFromLabels(indicies, testData)    
plotClusters(testPoints1, testPoints2, testPoints3, ax1, "v", 400, "Test Data ")

plt.show()
```

### 4. ###
Finally, we want you to deliberately find a non-optimal clustering. To this end, use the within-cluster sums of point-to-centroid distances returned by K-means to keep calling K-means until you find a clearly non-optimal value. You may need to change the definition of `kmeans` function and play around with its initialisation parameters. Plot the Voronoi diagram in the same plot to demonstrate non-optimality.

```python
noOfDataPoints, dimensions = data.shape

def kmeans2(Data, NClusters, centroids):
  km = KMeans(NClusters, n_init = 1, max_iter = 3, init=centroids)
  fitted = km.fit(Data)
  return (fitted.cluster_centers_, fitted.labels_, fitted.inertia_)

def nonOptimalClustering():
    centroids = np.array([[3,  3], [3,7], [7,7]])
    for i in range(4):
        cluster_centres, labels, inertia = kmeans2(data, 3, centroids)
        offsetVal = inertia / noOfDataPoints
        offset = np.repeat(offsetVal,6).reshape(3,2)
        centroids = np.add(centroids, offset)
    return cluster_centres, labels
    
noCluster_Centres, noLables = nonOptimalClustering()

vor = Voronoi(noCluster_Centres)
fig5 = voronoi_plot_2d(vor)
plt.title('Non optimal clustering of test Data')
ax5 = fig5.add_subplot( 111)

ax5.set_xlim(1, 10)
ax5.set_ylim(1, 10)

#Add cluster centres to plot
ax5.scatter(noCluster_Centres[:,0], noCluster_Centres[:,1], marker="+", s=900, label="Cluster Centroids")

ps1, ps2, ps3 = clusterFromLabels(noLables, data)    
plotClusters(ps1, ps2, ps3, ax5, "o", 100, "Training Data ")


```

The simple nearest-centroid classifier you constructed in the previous assignment leads to linear decision boundaries between each pair of classes, as visualised by the Voronoi diagram. We now want you to contrast and compare this with a maximum-likelihood classifier.

### 1. ###
Model the data in each class as being generated from a 2-D Normal Distribution. Estimate the class means and covariances from the labelled training data using the `np.mean` (remember about `axis` parameter) and `np.cov` functions. Visualise the estimated distributions by plotting contours of the `stats.multivariate_normal.pdf` functions of each class. Choose the contour level such that 95% of the probability mass is within the ellipse. Use as a starting point that for bivariate random distributions the points on this contour satisfy the equation
$$ (x-\mu)^{T} * \Sigma^{-1} * (x-\mu) / 2 = 3 $$
and use this to find the appropriate density for the contour level. `Python` functions to use include `plt.contour` and `np.meshgrid`.

**Clarification:** this basically tells you to draw an ellipse at squared Mahalanobis distance 6. Why 6 (actually 5.99)? This comes from the fact that sums of squared Gaussians follow a Chi-squared distribution, so we use the inverse of the Chi-squared cumulative distribution with 2 degrees of freedom. You can verify this in `Python` by the query `stats.chi2.ppf(0.95, 2)` which gives 5.9915. This is the 2-D equivalent of saying that in 1-D, 95% of the Gaussian probability mass is within +/- 2 standard deviations from the mean (actually 1.96). More information at the following links:  
* [http://www.visiondummy.com/2014/04/draw-error-ellipse-representing-covariance-matrix/](http://www.visiondummy.com/2014/04/draw-error-ellipse-representing-covariance-matrix/)
* [http://www.cs.ubc.ca/~murphyk/Teaching/CS340-Fall06/reading/gaussians.pdf](http://www.cs.ubc.ca/~murphyk/Teaching/CS340-Fall06/reading/gaussians.pdf)

```python
fig3 = plt.figure()
plt.title('Clustering with ellipsis including 95% of probability mass')

def ellipsisFunction(x,y, mu, covar):
    # Will calculate the following function when the inputs are provided (x)T1(x)/2=3
    covar = np.matrix(covar)
    xvector = np.array([x,y])
    a = (np.subtract(xvector , mu)).T
    b = covar.I
    c = (np.subtract(xvector , mu)) / 2
    c = np.matrix(c).T
    ans = (a * b * c).item(0)    
    return ans                       
    

clustercentres, labels, inertias = kmeans(data, 3)
trainData1, trainData2, trainData3 = clusterFromLabels(labels, data)

# N.B this pdf is not the one used in the question needs to be changed
def getEllipsis(x,y,mean,covar):
#   for each x and y in on the line it generates the Mahalonbis distance for each point on the graph
    values = [[ellipsisFunction(xi, yi, mean, covar)
            for xi in x] for yi in y]
    return values

def plotClusterContour(Data):
    x = np.linspace(0,10, num = 50)
    y = np.linspace(0,10, num = 50)
    
#   calculates the mean of the distribution which is a length 2 row vector
    mean = np.mean(Data[0], axis=0)
#   calculates the covariance of the distribution which is a 2x2 matrix
    covar = np.cov(Data[0], rowvar=False)
    
#   returns the squared Mahalanobis distance for each of the points x and y given a mean and covariance
    Z= getEllipsis(x,y,mean,covar)
    
# plots all the contours which a squared Mahalonbis distance of 6 on the graph
    plt.contour(x,y,Z,[3])


# for each cluster it plots the ellipsis such that 95% of the probability mass lie within the ellipsis
plotClusterContour(trainData1)
plotClusterContour(trainData2)
plotClusterContour(trainData3)

    
ax3 = fig3.add_subplot(111)
plotClusters(trainData1, trainData2, trainData3, ax3, "o", 100, "Training Data ")

```

### 2. ###
Now plot the decision boundaries between each pair of classes. This can be done once again using the `plt.contour` function, this time on each of the three pairwise likelihood ratios. Visually inspect the decision boundaries and make sure you understand why they have a particular shape. Also check whether this maximum-likelihood classifier disagrees with the nearest-centroid classifier on any of the test or training points (for both your data sets if you're working in pairs).

```python
# put your code here
x = np.linspace(0,10)
y = np.linspace(0,10)

def mvnPdf(x,y,mean,covar):
#   for each x and y in on the line it generates the probability that a sample is in the distribution
    pdf = [[stats.multivariate_normal.pdf((xi,yi),mean,covar) for xi in x] for yi in y]  
    return pdf

def getPdf(trainData):
    mean = np.mean(trainData, axis=0)
    covar = np.cov(trainData, rowvar=False)
    z= mvnPdf(x,y,mean,covar)
    return z
    
def plotClusterDifferences(z1,z2,z3):
    z4 = np.subtract(z1, z2)
    z5 = np.subtract(z2, z3)
    z6 = np.subtract(z1, z3)
    
    c1 = plt.contour(x,y, z4, [0], colors='black')
    c1.collections[0].set_label("Boundary between clusters 1 and 2")
    
    c2 = plt.contour(x,y, z5, [0], colors='g')
    c2.collections[0].set_label("Boundary between clusters 2 and 3")
    
    c3 = plt.contour(x,y, z6, [0], colors='r')
    c3.collections[0].set_label("Boundary between clusters 1 and 3")
    
    plt.legend(loc='lower right')
    
# for each cluster it plots the ellipsis such that 95% of the probability mass lie within the ellipsis

z1 = getPdf(trainData1[0])
z2 = getPdf(trainData2[0])
z3 = getPdf(trainData3[0])

# plots the decision boundaries between plots
vor = Voronoi(clustercentres)
fig3 = voronoi_plot_2d(vor)
# vor.set_label("simple nearest neighbour classifier")


# fig3 = plt.figure()
plt.title('Clustering with maximum likelihood decision boundaries')
plotClusterDifferences(z1,z2,z3)

plotClusterContour(trainData1)
plotClusterContour(trainData2)
plotClusterContour(trainData3)

ax3 = fig3.add_subplot(111)
ax3.set_xlim(1, 10)
ax3.set_ylim(1, 10)
ax3.scatter(clustercentres[:,0], clustercentres[:,1], marker="+", s=900, label="Cluster Centres")
plotClusters(testPoints1, testPoints2, testPoints3, ax3, "v", 400, "Test Data ")
plotClusters(trainData1, trainData2, trainData3, ax3, "o", 100, "Train Data ")
```

```python
def getPdfNoCovar(trainData):
    mean = np.mean(trainData, axis=0)
    covar = np.array([[1,0],[0,1]])
    z= mvnPdf(x,y,mean,covar)
    return z
    
zNoCovar1 = getPdfNoCovar(trainData1[0])
zNoCovar2 = getPdfNoCovar(trainData2[0])
zNoCovar3 = getPdfNoCovar(trainData3[0])

vor = Voronoi(clustercentres)
fig6 = voronoi_plot_2d(vor)
plt.title('Clustering with nearest centroid decision boundaries')
ax6 = fig6.add_subplot( 111)

ax6.set_xlim(1, 10)
ax6.set_ylim(1, 10)


ax6.scatter(clustercentres[:,0], clustercentres[:,1], marker="+", s=900, label = "Cluster Centres")
plotClusters(testPoints1, testPoints2, testPoints3, ax6, "v", 400, "Test Data ")
plotClusters(trainData1, trainData2, trainData3, ax6, "o", 100, "Train Data ")

plotClusterDifferences(zNoCovar1,zNoCovar2,zNoCovar3)
    
def getPdfMAP(trainData ,p):
    mean = np.mean(trainData, axis=0)
    covar = np.cov(trainData, rowvar=False)
    z= mvnPdf(x,y,mean,covar)
    z = p * np.array(z)
    #print(z)
    return z

fig7 = plt.figure()
plt.title('Clustering with MAP decision boundaries')
ax7 = fig7.add_subplot( 111)

def plotMAPEstimation(p1, p2, p3):
    z1 = getPdfMAP(trainData1[0], p1)
    z2 = getPdfMAP(trainData2[0], p2)
    z3 = getPdfMAP(trainData3[0], p3)
   
    plotClusterDifferences(z1,z2,z3)

plotMAPEstimation(0.5, 0.25, 0.25)    

ax7.scatter(clustercentres[:,0], clustercentres[:,1], marker="+", s=900, label = "Cluster Centres")
plotClusters(testPoints1, testPoints2, testPoints3, ax7, "v", 400, "Test Data ")
plotClusters(trainData1, trainData2, trainData3, ax7, "o", 100, "Train Data ")

    
```
